```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
x=np.array(Image.open('img04g.tif'))
plt.imshow(x,plt.cm.gray)
plt.title('clean image')
plt.show()
```

```python
y = x + 16 * x.std() * np.random.random(x.shape)
plt.imshow(y,plt.cm.gray)
plt.title('noisy image')
plt.show()
```

```python
[m,n]=x.shape
```

```python
K=20
N=m*n
sigmaw=16
```

```python
xnew=np.zeros([m,n])
xs=y
for f in range(K):
    for i in range(1,m-1):
        for j in range(1,n-1):
            sigmax=1.0/N*((xs[i,j]-x[i-1,j])**2+(xs[i,j]-xs[i+1,j])**2+(xs[i,j]-xs[i,j+1])**2+(xs[i,j]-xs[i,j+1])**2)
            sum_value=xs[i,j-1]+xs[i,j+1]+xs[i-1,j]+xs[i+1,j]
            numerator=y[i,j]+1.0*sigmaw**2/(sigmax)*(sum_value)
            denumerator=1+1.0*sigmaw**2/sigmax
            allvalue=numerator*1.0/denumerator
            xnew[i,j]=np.maximum(0,allvalue)
    xs=xnew/np.max(xnew)
            
    
```

```python
myall=xs
plt.imshow(myall,plt.cm.gray)
```
